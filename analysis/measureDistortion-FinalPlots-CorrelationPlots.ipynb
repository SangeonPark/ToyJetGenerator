```python
def sigmascore(pairwise_ratio):
    return np.var(pairwise_ratio/np.mean(pairwise_ratio))
```

```python
def bin_and_average(pairwise_ratio, original_metric, binning):
    mean_arr = []
    sigmascore_arr = []
    midpoints = []
    for left_bin, right_bin in zip(binning[:-1], binning[1:]):
        midpoints.append((left_bin+right_bin)/2.)
        #print(np.where((original_metric>left_bin)&(original_metric<right_bin))[0].size)
        #print(pairwise_ratio[np.where((original_metric>left_bin)&(original_metric>right_bin))])
        #print(np.mean(pairwise_ratio[np.where((original_metric>left_bin)&(original_metric<right_bin))]))
        #print(sigmascore(pairwise_ratio[np.where((original_metric>left_bin)&(original_metric<right_bin))]))
        mean_arr.append(np.mean(pairwise_ratio[np.where((original_metric>left_bin)&(original_metric<right_bin))]))
        #print(mean_arr)
        sigmascore_arr.append(sigmascore(pairwise_ratio[np.where((original_metric>left_bin)&(original_metric<right_bin))]))
    return mean_arr, sigmascore_arr, midpoints
```
