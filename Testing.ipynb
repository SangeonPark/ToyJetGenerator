```python
from datagenerator import jet_data_generator 
from plotutils import plot_event 
import matplotlib 
import time
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (6,6)
plt.rcParams.update({'font.size': 14})
```

```python
sig_2p_8part = jet_data_generator("signal",0.00005, 2, 8)
```

```python
start = time.time()
data_sig_2p_8part = sig_2p_8part.generate_dataset(1000)
end = time.time()
print(end-start)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.hist(sig_2p_8part.zhard,alpha=0.2,density=True,label="Hard Splitting")
ax.hist(sig_2p_8part.zsoft,alpha=0.2,density=True,label="Soft Splitting")
ax.legend()
ax.set_xlabel(r'$z_{g}$')
ax.set_ylabel('Density')
fig.savefig('./figures/momentum_sharing.png')
```

```python
def plot_event(pfcands, name):

    pt = pfcands[:,0]
    eta = pfcands[:,1]
    phi = pfcands[:,2]
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    ax.scatter(eta,phi,s=pt*200, alpha=0.2)
    ax.set_xlabel('eta')
    ax.set_ylabel('phi')
    ax.set_xlim([-4,4])
    ax.set_ylim([-4,4])
    fig.savefig(f'figures/{name}.png')
```

```python
plot_event(data_sig_2p_8part[2].reshape(-1,3), "sig_2p_8part")
```
